```python
import pandas as pd
import tsanomaly as ad
```

```python
df = pd.read_csv("./data/DailyDelhiClimateTrain.csv").set_index("date")
df.index = pd.to_datetime(df.index)
df.head()
```

```python
x = df.loc["2013-01-01":"2013-06-30", "meantemp"]
ewma_result = ad.ewma(x, 0.8, 3.0)
ad.plot_anomalies(ewma_result)
```
